```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import time
from selenium.common.exceptions import StaleElementReferenceException
```

```python
driver = webdriver.Chrome(executable_path="C:/Users/nwadhwani/OneDrive - Navisite/Desktop/chromedriver")
driver.maximize_window()
driver.implicitly_wait(7)
driver.get("https://www.sanjeevkapoor.com/AllRecipes")
list_links = []
receipes = driver.find_elements_by_class_name("image-thumbnail")
for rec in receipes:
    link = rec.find_element_by_tag_name("img").get_attribute("src")
    print(link)
loadmore = driver.find_element_by_id("ctl00_ContentPlaceHolder1_btnmore")
j = 0
try:
    while loadmore.is_displayed():
        loadmore.click()
        time.sleep(10)
        lrec = driver.find_elements_by_class_name("image-thumbnail")
        newlist = lrec[j:]
        for rec in newlist:
            link = rec.find_element_by_tag_name("img").get_attribute("src")
            print(link)
            list_links.append(link)
        j = len(lrec)+1
        time.sleep(7)
except StaleElementReferenceException:
    pass
driver.quit()
print(list_links)
```
